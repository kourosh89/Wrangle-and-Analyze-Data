## Introduction

In this project I wrangled and analyzed data from a Twitter account called WeRateDogs. WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog. These ratings almost always have a denominator of 10. The numerators, though? Almost always greater than 10. 11/10, 12/10, 13/10, etc. Why? Because "they're good dogs Brent." WeRateDogs has over 4 million followers and has received international media coverage.
The objective of this project was to clean the gathered data from twitter account and finally derive interesting conclusions from the results. Some questions that could be answered. Which dog breed has the highest rating? Which dog stage is the most frequent? What is the relationship betweet favourite and retweet counts vs dog rating?

## Which Dog Breeds and Stages had the most frequency in tweets

we had some various stages of dogs that people mention them in tweeter. we analized them and filled out some blank breed using text column to figure out which breed belonged to those rows. finally we collected data about 4 breeds that 'pupper' breed has maximum breeds in our dataset. also we discovered that 'floofer' breed tweeted at the minimum count

```python
import pandas as pd, seaborn as sb, matplotlib.pyplot as plt
plt.figure(figsize=[9,4])
df=pd.read_csv('twitter_archive_master.csv')
color_base=sb.color_palette()[0]
sb.countplot(data=df , x='breed' , color=color_base,order=df.breed.value_counts().index)
plt.title('number of each breed in tweets')
plt.show()
```

## Which score have the maximum value in tweets

we had rates people's dogs with a humorous comment about the dog. These ratings almost always have a denominator of 10. The numerators are greater than 10. the only factor that can help us to know which numerator and denominator totally have maximum value, using th rate that is name 'score'. we found the maximum rate of dog score happened in 1.2 it shows that majority of tweets would prefer to give rating_numerator close to rating_denominator. high score of rating_numerator occured less 

```python
color_base=sb.color_palette()[1]
plt.figure(figsize=[16,5])
sb.countplot(data=df , x='score' , color=color_base,order=df.score.value_counts().index)
plt.title('number of score in tweets')
plt.show()
```

## Which breed has the maimum tweet in top three score

this is really good report to know about breeds. we got 13 categoris of 13 score. but we analized for top three of them. we understood that 'pupper' has maximum value in each score. we found 'puppo' in rating 1 and 1.2 are more than floofer, while in score 1.1 they are approximately same.     

```python
import warnings
warnings.filterwarnings('ignore')
df1=df.query('score==1.2 or score==1.1 or score==1')
df1.score=df1.score.astype("str")
plt.figure(figsize=[16,5])
sb.countplot(data=df1 , x='score' , hue='breed')
plt.title('comparision of score against breed')
plt.show()
```

## What is the relationship between score and favorite_count and followers_count

in this visualization, we discovered that majority of 'favorite_count' happend in score 1
we have some favorite_count for tweet without score. for 'followers_count' we have the same situation. but the number of followers_count in zero score happend in high scale

```python
df['score']=df['score'].astype("int")
fig , ax = plt.subplots(ncols=2 , figsize=[16,5])
sb.regplot(data=df,x="score" , y='favorite_count',fit_reg=False,ax=ax[0])
ax[0].set_title('favorite_count against varieties of score')
sb.regplot(data=df,x="score" , y='followers_count',fit_reg=False,ax=ax[1])
ax[1].set_title('followers_count against varieties of score')
plt.show()
```
